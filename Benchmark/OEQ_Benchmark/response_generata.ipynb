```python
import pandas as pd

vh_mode_list = ["counting", "shape", "color", "orientation", "size", "position", "OCR", "existence"]

for task in vh_mode_list:
    df = pd.read_excel(f'{task}.xlsx')[['Question', 'Reference_Answer', 'raw_response']]

    with open(f'Raw_{task}.txt', 'w') as file:
        for index, row in df.iterrows():
            file.write(f"Question: {row['Question']}, answer: {row['Reference_Answer']}, response: {row['raw_response']}\n")
```
