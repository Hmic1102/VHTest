```python
import numpy as np
import pandas as pd
import re
```

```python
task = 'existence'
```

```python
file_path = 'output.txt'

with open(file_path, 'r') as file:

    lines = file.readlines()


data = []

for i in range(0, len(lines), 2):  # Two lines per entry

    response = lines[i].strip()

    index_line = lines[i+1].strip()

    index = int(index_line.split(":")[1].strip().split(",")[0])

    data.append({"index": index, "response": response})



df = pd.DataFrame(data)
task_df = pd.read_excel(f'{task}.xlsx')
task_df['raw_response'] = df.response
```
